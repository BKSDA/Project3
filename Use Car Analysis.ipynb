```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import isnan
import seaborn as sns
```

```python
#Importing dataset for analysis
df_user_car_data = pd.read_csv('UserCarData.csv')
```

```python
#Checking importation was successful
df_user_car_data
```

```python
#Check the shape for number of columns and rows
df_user_car_data.shape
```

```python
#Check data types for each column
df_user_car_data.info()
```

```python
#Obtain basic statistical measures
df_user_car_data.describe()
```

```python
#Question 1= How many cars were sold each year?
Q1_number_cars_sold_per_year = df_user_car_data.groupby(
    ['year']).count()['Sales_ID']
Q1_number_cars_sold_per_year.plot(kind='bar')
```

```python
#ANSWER1: 2017 was the year with most cars sold.
```

```python
#Question2= Which year had the most total sales?
Q2_sales_total_per_year = df_user_car_data.groupby(
    ['year']).sum()['selling_price']
Q2_sales_total_per_year
```

```python
Q2_sales_total_per_year = df_user_car_data.groupby(
    ['year']).sum()['selling_price']
Q2_sales_total_per_year.plot(kind='bar')
```

```python
#ANSWER2= 2019 was the year with more total sales.
```

```python
#Question3= Which state sold most cars per year?
sales_per_year_state = df_user_car_data.groupby(
    ['year','State or Province']).count()['Sales_ID'].unstack()
sales_per_year_state
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(sales_per_year_state,
            cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
#ANSWER3 = California has been selling more cars from 2010 till 2020.
```

```python
# Question4 = Which cars are more popular for selling?
name_sales_count_per_year = df_user_car_data.groupby(
    ['year','name']).count()['Sales_ID'].unstack()
name_sales_count_per_year
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(name_sales_count_per_year,
            cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
#ANSWER4: Maruti are the most popular cars for selling.
```

```python
#Question5= Which cars sold generate higher sales prices?
name_sales_amount_per_year = df_user_car_data.groupby(
    ['year','name']).sum()['selling_price'].unstack()
name_sales_amount_per_year
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(name_sales_amount_per_year,  
            cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
#ANSWER5 = BMWs in 2019, Jaguar in 2017, Lexus second best in 2019.
```

```python
#QUESTION6= What portion of the market belongs to each region in relation to sales amount? 
region_sales = df_user_car_data.groupby(
    ['Region']).sum()['selling_price']
region_sales
```

```python
region_sales.plot.pie(ylabel= '' , title="Total Sales per Region", legend=False,
                   autopct='%1.1f%%')
```

```python
#ANSWER6 = The central region has the highest portion of the market,
#yet, the the % is distributed evenly through the whole country.
```

```python
#Question7= How many owners do cars have prior to selling based on name of car?
plt.figure(figsize=(20, 10))
sns.scatterplot(x= 'owner',
                y= 'name',
                hue='year',
                data=df_user_car_data)
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x= 'km_driven',
                y= 'owner',
                hue='transmission',
                data=df_user_car_data)
```

```python
#Comparing multiple variables with seaborn
plt.figure(figsize=(50, 10))
sns.pairplot(df_user_car_data, vars=[
    'km_driven','year', 'name', 'owner'])
plt.show()
```

```python
owner_numberofsales = df_user_car_data.query(
    "owner != 'First_Owner'").groupby(
    "owner")['Sales_ID'].agg(['mean','count']).query(
        'count >= 100')
display(owner_numberofsales)
owner_numberofsales.plot()
```

```python
maruti_sales = df_user_car_data.query("name == 'Maruti'").groupby(
    "year")['selling_price'].agg(['mean','sum']).query(
        'sum >= 500000')
display(maruti_sales)
maruti_sales.plot()
```
